```python
import pandas as pd
dt=pd.DataFrame({'x':[4,5,8,9],'y':[8,9,6,7]},index=['a','b','c','d'])
dt['x'].plot(kind='pie',subplots=True,labels=dt.index,figsize=(7,7),autopct='%1.1f%%')
dt
```

```python
#READ DATA FROM EXCEL FILE INTO COLLAB AND PRINT BAR CHART AND PIE CHART AND TAB

import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('bill.csv')
df=pd.DataFrame(data)

# Plot a bar chart
df.plot(kind='bar', x='DAYS', y='FEE')

# Plot a pie chart
df.plot(kind='pie',subplots=True,labels=df.index, figsize=(8,8), autopct='%1.1f%%')
df
```

```python
import pandas as pd
df=pd.DataFrame({'teams':['a','a','b','b','c','c'],'values':[15,16,17,46,53,63]})
df.groupby(['teams']).sum().plot(kind='pie',y='values',autopct='%1.1f%%',title='points obtained')
df

```
