````python
!pip install openpyxl pandas matplotlib
import pandas as pd
import io
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import files

# Uploading files
uploaded = files.upload()
# Read the first Excel file for semester 1
df_sem1 = pd.read_excel(uploaded[list(uploaded.keys())[0]])

# Read the second Excel file for semester 2
df_sem2 = pd.read_excel(uploaded[list(uploaded.keys())[1]])
# Merge the two semester dataframes based on 'Student' column
df_combined = pd.merge(df_sem1, df_sem2, on='Student', suffixes=('_Sem1', '_Sem2'))

# Calculate total marks obtained and total marks possible for each student
df_combined['Total_Marks_Obtained'] = df_combined['Marks_Obtained_Sem1'] + df_combined['Marks_Obtained_Sem2']
df_combined['Total_Possible_Marks'] = df_combined['Total_Marks_Sem1'] + df_combined['Total_Marks_Sem2']

# Calculate the percentage for each student
df_combined['Percentage'] = (df_combined['Total_Marks_Obtained'] / df_combined['Total_Possible_Marks']) * 100

print(df_combined[['Student', 'Total_Marks_Obtained', 'Total_Possible_Marks', 'Percentage']])

# Plotting the pie chart of total marks obtained by each student
labels = df_combined['Student']  # Labels are the names of the students
sizes = df_combined['Total_Marks_Obtained']  # The sizes of the pie are the total marks obtained by each student

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Percentage of Marks Obtained by Each Student")
plt.show()
````

### REGRESSION PLOT

````python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2,4,6,4,5])

# Step 1: Calculate the mean of X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Step 2: Calculate the slope using the formula
numerator = np.sum((X - mean_x) * (Y - mean_y))
denominator = np.sum((X - mean_x) ** 2)
slope = numerator / denominator

# Step 3: Calculate the intercept
intercept = mean_y - (slope * mean_x)

# Step 4: Predict Y values based on the regression line
Y_pred = slope * X + intercept

# Step 5: Calculate R^2 (coefficient of determination)
ss_total = np.sum((Y - mean_y) ** 2)  # Total sum of squares
ss_residual = np.sum((Y - Y_pred) ** 2)  # Residual sum of squares
r_squared = 1 - (ss_residual / ss_total)

# Step 6: Display the slope, intercept, and R^2
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R^2: {r_squared:.2f}")

for i in range(len(X)):
    plt.text(X[i], Y[i], f'({X[i]}, {Y[i]})', fontsize=6, ha='right')



# Step 7: Plot the data and regression line
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label=f'y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title(f'Regression Line with R^2 = {r_squared:.2f}')
plt.legend()
plt.show()
````

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Example data
years = pd.date_range(start='2000-3-10', end='2021-06-10', periods=5).strftime('%d %b %Y').tolist()
coal = [10, 15, 17, 16, 43]
wind = [12, 14, 61, 23, 42]
solar = [21, 24, 23, 21, 20]

# Total for each category
total_coal = np.sum(coal)
total_wind = np.sum(wind)
total_solar = np.sum(solar)

# Calculate percentages
coal_percent = [(x / total_coal) * 100 for x in coal]
wind_percent = [(x / total_wind) * 100 for x in wind]
solar_percent = [(x / total_solar) * 100 for x in solar]

# Scatter plot for each category
plt.stackplot(years,solar_percent,wind_percent,coal_percent, labels=['solar_percent','wind_percent','coal_percent'],colors=['maroon','green','yellow'])

# Add percentage labels to each point
for i in range(len(years)):
    plt.text(years[i], solar[i], f'{solar_percent[i]:.1f}%', ha='center', va='bottom', fontsize=8, color='maroon')# horizontal alignment
    plt.text(years[i], wind[i], f'{wind_percent[i]:.1f}%', ha='center', va='bottom', fontsize=8, color='green')#vertical alignment
    plt.text(years[i], coal[i], f'{coal_percent[i]:.1f}%', ha='center', va='bottom', fontsize=8, color='yellow')


# Customize the plot
plt.title('Energy Consumption Over Time with Percentages')
plt.xlabel('Year')
plt.ylabel('Energy')
plt.legend(loc='upper left')

# Display the plot
plt.show()
````

````python
!pip install openpyxl
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Upload file
uploaded = files.upload()

# Load data
rate = pd.read_excel(uploaded[list(uploaded.keys())[0]], index_col=0)
print(rate)

# Line Plot
# rate.index: The index of the DataFrame represents the categories (such as age groups) in each year. It is used as the x-axis values for the plot.
# rate[col]: This extracts the data for the current column (i.e., the percentages for the age groups in that year), and these are plotted on the y-axis.
# marker='o': This adds circular markers to each data point on the line plot.
# label=col: This sets the label for the legend, using the name of the current column (the year).

plt.figure(figsize=(10, 6))
for col in rate.columns:
    plt.plot(rate.index, rate[col], marker='o', label=col)
plt.title('Percentage Distribution Over Years')
plt.xlabel('Age group')
plt.ylabel('Percentage')
plt.legend(title='Year')
plt.grid(True)
plt.show()

# Bar Plot
rate.plot(kind='bar', figsize=(10, 6), colormap='viridis')
plt.title('Stacked Bar Plot of Age Group Percentage Distribution Over Years')
plt.xlabel('Age group')
plt.ylabel('Percentage')
plt.legend(title='Year')
plt.show()

# Heatmap (Transpose the DataFrame for better representation)
plt.figure(figsize=(8, 6))
sns.heatmap(rate, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Age Group Percentage Distribution Over Years')
plt.xlabel('Age group')
plt.ylabel('Year')
plt.show()

# Colormap (also transposed for readability)
plt.figure(figsize=(8, 6))
sns.heatmap(rate, annot=True, cmap='Blues', cbar=True)
plt.title('Colormap of Age Group Percentage Distribution Over Years')
plt.xlabel('Age group')
plt.ylabel('Year')
plt.show()

````

try different types of imputataions in pandas for finding the missing values (mean ,medina etccc)





````python
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Region': ['North', 'South', 'East', 'West', 'North'],
    'Revenue': [250, np.nan, 150, 200, np.nan]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
df_mean_imputed = df.copy()
df_mean_imputed['Revenue'].fillna(df['Revenue'].mean(), inplace=True)
print("\nDataFrame with Mean Imputation:")
print(df_mean_imputed)
df_median_imputed = df.copy()
df_median_imputed['Revenue'].fillna(df['Revenue'].median(), inplace=True)
print("\nDataFrame with Median Imputation:")
print(df_median_imputed)
df_mode_imputed = df.copy()
mode_value = df['Revenue'].mode()[0]
df_mode_imputed['Revenue'].fillna(mode_value, inplace=True)
print("\nDataFrame with Mode Imputation:")
print(df_mode_imputed)
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
import shap
from google.colab import files
uploaded = files.upload()
# Load dataset using pd.read_excel since it's an Excel file (.xlsx)
df = pd.read_excel("churn_data.xlsx") # Corrected filename based on the Global Variables

# Step 1: Initial Exploration
print(df.head())
print(df.info())
print(df.isnull().sum())
print(df.describe())
df[['Age', 'MonthlyCharges', 'Tenure', 'TotalCharges']].hist(bins=20, figsize=(10, 8))
plt.show()

# Step 2: Handle Missing Values and Encode Categorical Variables
imputer = SimpleImputer(strategy='mean')
df['TotalCharges'] = imputer.fit_transform(df[['TotalCharges']])

# One-hot encode categorical columns
df = pd.get_dummies(df, columns=['Gender', 'InternetService', 'ContractType', 'PaymentMethod'], drop_first=True)

# Step 3: Feature Engineering
df['LTV'] = df['MonthlyCharges'] * df['Tenure']
df['EngagementScore'] = df[['StreamingTV', 'StreamingMovies', 'InternetService_Fiber optic', 'DeviceProtection']].sum(axis=1)
df['TenureGroup'] = pd.cut(df['Tenure'], bins=[0, 12, 24, df['Tenure'].max()], labels=['New', 'Mid', 'LongTerm'])

# Step 4: EDA
sns.countplot(x='Churn', hue='Gender_Male', data=df)
plt.title("Churn by Gender")
plt.show()

sns.histplot(df[df['Churn'] == 1]['Tenure'], color='red', label='Churned', kde=True)
sns.histplot(df[df['Churn'] == 0]['Tenure'], color='blue', label='Not Churned', kde=True)
plt.title("Tenure Distribution")
plt.legend()
plt.show()

# Select only numeric features for correlation analysis
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm') # Use numeric_df instead of df
plt.title("Correlation Matrix")
plt.show()

# Step 5: Model Building and Evaluation
X = df.drop(columns=['Churn', 'CustomerID', 'TenureGroup']) # Drop 'TenureGroup'
y = df['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("AUC Score:", roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]))

#Step 6: Model Interpretation with SHAP
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test)

# Select SHAP values for the positive class (churn = 1)
# Change 1: Indexing to select SHAP values for class 1 correctly
shap_values_class1 = shap_values[1]

# Change 2: Pass the correct SHAP values to summary_plot
shap.summary_plot(shap_values_class1, X_test)

# Step 7: Prediction Function for New Data
def predict_churn(new_data):
    prediction = model.predict(new_data)
````
