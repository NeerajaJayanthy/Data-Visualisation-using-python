```python
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
upload=files.upload()

#QN 1
data = pd.read_csv('World_Happiness_2023.csv')
print(data.head())

#QN 2
sorted_data =data.sort_values(by='Happiness_Score', ascending=False)
top_10_happiest = sorted_data.head(10)
print("\nthe top 10 countries\n",top_10_happiest)

#QN 3
sorted_data =data.sort_values(by=['Region','Happiness_Score'], ascending=[True,False])
happiest = sorted_data.head(15)
print("\nthe top 15 country\n",happiest)

#QN 4
filtered=data[(data['Happiness_Score']>7)&(data['GDP_Per_Capita']>1)]
print("\nthe filtered data\n",filtered)

#QN 5
filtered_regions = data[data['Region'].isin(['Western Europe', 'North America'])]
print(filtered_regions)

#QN 6
freedom_filtered = data[data['Freedom'].between(0.5, 1.0)]
print(freedom_filtered)

#QN 7
data['Economic_Factors'] =data['GDP_Per_Capita'] + data['Social_Support']
nan_rows = data[data['Economic_Factors'].isna()]
print(nan_rows)

#QN 8
cleaned_data =data.dropna(subset=['Corruption'])
print(f"Number of remaining rows: {cleaned_data.shape[0]}")

#QN 9
data['Happiness_Category'] = pd.cut(data['Happiness_Score'],bins=[-float('inf'), 5, 7, float('inf')],labels=['Low', 'Medium', 'High'])
print(data)

#QN 10
fig, ax = plt.subplots(figsize=(10, 6))
top_10_happiest[['Happiness_Score', 'GDP_Per_Capita']].plot(kind='bar', ax=ax)
ax.set_ylabel('Scores')
ax.set_title('Top 10 Happiest Countries: Happiness Score and GDP Per Capita')
plt.tight_layout()
plt.show()

#QN 11
filtered_countries = data[(data['Region'].isin(['Western Europe', 'North America'])) &(data['Freedom'] > 0.8) &(data['Corruption'] < 0.3)]
sorted_filtered_countries = filtered_countries.sort_values(by='GDP_Per_Capita', ascending=False)
top_5_countries = sorted_filtered_countries[['Happiness_Score', 'Freedom']].head(5)
print(top_5_countries)
```
